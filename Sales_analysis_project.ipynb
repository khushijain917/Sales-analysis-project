```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"],dayfirst=True)
```

```python

# adding Extra columns for analysis
df['Month'] = df['Order_Date'].dt.to_period('M')
df['Profit'] = df['Total_Sales'] - df['Total_Cost']
df['ASP'] = df['Total_Sales'] / df['Quantity']
```

```python
df.dtypes
```

```python
# Prepare the data first
monthly_profit = df.groupby("Month")["Profit"].sum()

# Now plot the data
plt.figure(figsize=(15, 5))
ax = monthly_profit.plot(kind="line", marker='o', color='blue')

# Add value labels on each point
for x, y in zip(monthly_profit.index.astype(str), monthly_profit):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=9)

plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Sales Trade Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()

```
